```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import chi2
```

```python
df=pd.read_csv("/home/umrn5505/Desktop/phisingDetection/Phishing_detection/datasets/url_features.csv")
```

```python
type(df["symbol_at_end"][0])
features_to_convert= df.select_dtypes(include=bool).columns
df[features_to_convert]=df[features_to_convert].astype(int)
df
```

```python
X= df.drop(columns=["type","url"])
Y=df["type"].map(
    {
        "legitimate": 0,
        "phishing": 1
    }
)
print(X.head())
print(Y.head())
```

```python
Y.unique()
```

```python
chi2_stats,p_values=chi2(X,Y)
```

```python
chi2_df = pd.DataFrame({
    "feature": X.columns,
    "chi2_score": chi2_stats,
    "p_value": p_values
})
chi2_df = chi2_df.sort_values(by="chi2_score", ascending=False)
```

```python
chi2_df
```

```python
from sklearn.feature_selection import SelectKBest

selector = SelectKBest(score_func=chi2, k=9)
selector.fit(X, Y)

selected_features = X.columns[selector.get_support()]
print(selected_features)
```
